```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def sampling(img,n_x,n_y):
    r,c=img.shape
    n_x=round(r/n_x)
    n_y=round(c/n_y)
    new_img=[]
    num=-1
    for i in range(0,r,n_x):
        new_img.append([])
        num+=1
        for j in range(0,c,n_y):
            new_img[num].append(img[i,j])
    return np.array(new_img)
```

```python
im = cv2.imread("../input_data/bricks.jpg",0)
r,c=im.shape

n = range(2,11)

fig = plt.figure(figsize = (10,5))
plt.subplot(2,5,1)
plt.imshow(im,cmap='gray')
plt.title("Original")

for i in n:
    out = sampling(im,r//i,c//i)
    plt.subplot(2,5,i)
    plt.imshow(out,cmap='gray')
    temp = "nx = "+str(r//i)+" ny = "+str(c//i)
    plt.title(temp)
    
plt.show()
```

```python
im2 = cv2.GaussianBlur(im,(15,15),0)
r,c=im2.shape

n = range(2,11)

fig = plt.figure(figsize = (10,5))
plt.subplot(2,5,1)
plt.imshow(im2,cmap='gray')
plt.title("Blurred")

for i in n:
    out = sampling(im2,r//i,c//i)
    plt.subplot(2,5,i)
    plt.imshow(out,cmap='gray')
    temp = "nx = "+str(r//i)+" ny = "+str(c//i)
    plt.title(temp)
    
plt.show()
```
